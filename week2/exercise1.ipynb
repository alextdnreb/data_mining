# Read Data

```python
import pandas as pd

dataset = pd.read_excel("CustomerDataSet.xls")

dataset
```

# Apply Clustering

```python
from sklearn.cluster import KMeans


estimator = KMeans(n_clusters=3)
result = estimator.fit_predict(dataset[["ItemsBought", "ItemsReturned"]])
dataset["cluster"] = result
dataset.sort_values(by="cluster", inplace=True)
print(dataset)
```

# Print Clusters

```python
import matplotlib.pyplot as plt

plt.title = "test"

# set the axis labels
plt.xlabel('ItemsBought')
plt.ylabel('ItemsReturned')

plt.scatter(dataset["ItemsBought"], dataset["ItemsReturned"], c=dataset["cluster"])

plt.show()
```

# Hierarchical Clustering

```python
from scipy.cluster.hierarchy import linkage, dendrogram

clusters = linkage(dataset[["ItemsBought", "ItemsReturned"]], "complete")
print(clusters)
# set the plot title
plt.title = 'Hierarchical Clustering Dendrogram'

# set the axis labels
plt.xlabel('Customer IDs')
plt.ylabel('distance')

# plot the dendrogram
dendrogram(clusters, labels=list(dataset['Customer ID']))

plt.show()
```
